```python
import numpy as np
import matplotlib.pyplot as plt
#loads in the dataset as a 82-length list, each element of which is a tuple of length 3 containing
#the three column values for that row
data = np.genfromtxt("ASTR19_F24_group_project_data.txt",
                     dtype=[('day','i8'),('time','U6'),('height','f8')])

#setting empty lists I will fill up with the column data
day = []
time = []
height = []

#print(data)
print(data.shape)

#looping over every row of the data
for i in range(82):
    #assigning first column to the day variable
    day.append(data[i][0])
    #assigning second column to the time variable
    time.append(data[i][1])
    #assigning third column to the height variable
    height.append(data[i][2])
    
print(day)
print(' ')
print(time)
print(' ')
print(height)
```

```python
plt.plot(height)
```

```python
def time_convert(day, time):
    hours = time[:time.index(":")]
    
    minutes = time[time.index(":")+1:]
    
    minutes_fl = float(minutes)/60
    
    hours_fl = (float(hours) + minutes_fl)/24
    
    return day + hours_fl
```

```python
Day = []
for i in range(len(time)):
    Day.append(time_convert(day[i], time[i]))
    
plt.plot(Day, height, '-o')
plt.ylabel("Height(ft)")
plt.xlabel("Day")
```

```python
print(Day)
```

```python
import tensorflow as tf
```

```python
Amp_init = 4.0
Amp_fit = tf.Variable(Amp_init, name='amp')

AmpFreq_init = 11.0
AmpFreq_fit = tf.Variable(AmpFreq_init, name='amp_freq')

AmpPh_init = 0.0
AmpPh_fit = tf.Variable(AmpPh_init, name='amp_phase')

TideFreq_init = 0.2
TideFreq_fit = tf.Variable(TideFreq_init, name='tide_freq')

TidePh_init = 1.5
TidePh_fit = tf.Variable(TidePh_init, name='tide_phase')

Offset_init = 2.478
Offset_fit = tf.Variable(Offset_init, name='offset')
```

```python
@tf.function
def model(x):
    return 4*tf.math.cos(AmpFreq_fit*x + AmpPh_fit) * tf.math.sin(TideFreq_fit*x + TidePh_fit) + Offset_fit
```

```python
ypred = model(Day)

plt.plot(Day, height, '-o')
plt.plot(Day, ypred)
plt.ylabel("Height")
plt.xlabel("Day")
```

```python
print(tf.__version__)
```

```python
@tf.function
def mse(ytrue, ypred):
    return tf.losses.mean_squared_error(ytrue, ypred)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
```

Amp_init = 4.0

AmpFreq_init = 20.0
AmpFreq_fit = tf.Variable(AmpFreq?_init, name='amp_freq')

AmpPh_init = 2.0

TideFreq_init = 1.5
TideFreq_fit = tf.Variable(TideFreq_init, name='tide_freq')

TidePh_init = 3.0
TidePh_fit = tf.Variable(TidePh_init, name='tide_phase')

Offset_init = 2.2
Offset_fit = tf.Variable(Offset_init, name='offset')


```python
epochs = 10000

display_step = 1000

for epoch in range(epochs):
    
    if ((epoch%display_step)==0):
        pred = model(Day)
        loss = mse(height, pred)
        
        print(f"Epoch {epoch} | Loss {loss.numpy()}, AF {AmpFreq_fit.numpy()} AP {AmpPh_fit.numpy()} TF {TideFreq_fit.numpy()} TP {TidePh_fit.numpy()} Offset {Offset_fit.numpy()}")
        
    with tf.GradientTape() as g:
        pred = model(Day)
        loss = mse(height, pred)
        
    gradients = g.gradient(loss, [AmpFreq_fit, AmpPh_fit, TideFreq_fit, TidePh_fit, Offset_fit])
    
    optimizer.apply_gradients(zip(gradients, [AmpFreq_fit, AmpPh_fit, TideFreq_fit, TidePh_fit, Offset_fit]))
    

print(f"Epoch {epoch} | Loss {loss.numpy()}, AF {AmpFreq_fit.numpy()} AP {AmpPh_fit.numpy()} TF {TideFreq_fit.numpy()} TP {TidePh_fit.numpy()} Offset {Offset_fit.numpy()}")
print("Done!")
```

```python
import sys
import scipy.special import erf

x=2.43

def event_probability(x,mu=0.0,s=1.0):
    z=np.fabs((x-mu)/s)
    def zfunc(z):
        return 0.5*(1.0+erf(z/2**0.5))
    return 1.0-(zfunc(z)-zfunc(-1*z))

if(len(sys.argv)>1):
    x=float(sys.argv[1])
    
prob=event_probability(x)

print(f"The Gaussian probability of evens larger than |{x}| is {prob*100:6.4f}%.")
print(f"The Gaussian probability of evens smaller than |{x}| is {(1-prob)*100:6.4f}%.")
```
